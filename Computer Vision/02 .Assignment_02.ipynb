#### 1. Explain convolutional neural network, and how does it work ?

A convolutional neural network (CNN) is a type of deep neural network that is commonly used for image recognition, natural language processing, and other tasks that involve processing data with a grid-like structure.

The basic building block of a CNN is the convolutional layer, which applies a set of filters (also called kernels or weights) to the input data. Each filter performs a convolution operation by sliding over the input data and computing the dot product between the filter and the overlapping region of the input at each position. The output of the convolution operation is a feature map that represents the presence or absence of certain features in the input.

The convolutional layer is typically followed by a non-linear activation function, such as the Rectified Linear Unit (ReLU), which applies an element-wise thresholding operation to the output of the convolutional layer. This allows the CNN to learn non-linear relationships between the input data and the desired output.

In addition to convolutional layers, CNNs also often include other types of layers, such as pooling layers and fully connected layers. Pooling layers downsample the feature maps by taking the maximum or average value within a local region of the feature map. This helps to reduce the spatial dimensions of the feature maps and increase the computational efficiency of the network. Fully connected layers connect every neuron in one layer to every neuron in the next layer, and are typically used for classification tasks where the output is a fixed number of classes.



CNNs are typically trained using a supervised learning approach, where the network is presented with labeled training data and adjusts its weights to minimize the difference between the predicted output and the true output. This is done using an optimization algorithm, such as stochastic gradient descent, to update the weights of the network based on the error between the predicted output and the true output.

#### 6. Explain definition of  receptive field ?


In the context of convolutional neural networks (CNNs), the receptive field is a term used to describe the portion of the input image that influences the activation of a particular neuron in the network. Specifically, the receptive field of a neuron in a CNN is defined as the area in the input image that is visible to that neuron through its connections to the previous layer of the network.

The receptive field of a neuron in a CNN is determined by the size of the kernel (also called filter or window) and the stride used in the previous layer, as well as the number and arrangement of the layers in the network. For example, a neuron in the second layer of a CNN with a kernel size of 3x3 and a stride of 1 has a receptive field of 3x3 pixels in the input image. If the same neuron is in the third layer of the network and connected to a previous layer with a kernel size of 3x3 and a stride of 2, then its receptive field would be larger, covering a 5x5 pixel area in the input image.
