ResNet (short for Residual Network) is a deep neural network architecture that is designed to address the problem of vanishing gradients and improve training of very deep networks. The core idea behind ResNet is the use of residual connections, which allow information to bypass one or more layers of the network and be directly propagated to subsequent layers.

Residual connections are implemented using shortcut connections, which connect the output of one layer to the input of a subsequent layer that is deeper in the network. These shortcut connections add an identity mapping to the output of the layer, which is combined with the output of the subsequent layer using element-wise addition. This effectively creates a "shortcut" path for the information to propagate through the network, allowing it to bypass any potentially problematic layers.

#### 7. What do Skip Connections entail ?


Skip connections, also known as shortcut connections, are a technique used in deep neural networks to improve the flow of information during training. They are used to connect the output of one layer directly to the input of a layer that is deeper in the network, effectively creating a "shortcut" path for information to propagate through the network. This allows the network to bypass certain layers that may be difficult to train, and helps to prevent the problem of vanishing gradients, which can occur in very deep networks.

Skip connections are implemented using element-wise addition, which combines the output of one layer with the output of a subsequent layer. This allows the network to learn residual mappings, which represent the difference between the current input and the desired output. By using these residual mappings, the network can effectively learn to "fill in the gaps" between the input and output, and can better capture the underlying relationships between the input features and the output.

Skip connections are a key component of many state-of-the-art neural network architectures, including ResNet, DenseNet, and U-Net. They have been shown to significantly improve the performance of these architectures on a wide range of computer vision tasks, including image classification, object detection, and segmentation.

#### 9. How can transfer learning help with problems ?



Transfer learning is a machine learning technique that involves using a pre-trained model on one task as a starting point for training a model on a different but related task. This approach can help with problems in several ways:

- Data efficiency: Pre-trained models are typically trained on large datasets, which allows them to capture general patterns that are useful for many different tasks. By starting with a pre-trained model, you can leverage this pre-existing knowledge to reduce the amount of data required to train a new model.

- Computational efficiency: Training deep neural networks from scratch can be computationally intensive, especially if you have limited hardware resources. By starting with a pre-trained model, you can avoid the need to train a large number of parameters from scratch, which can save a significant amount of time and resources.

- Performance: Pre-trained models have been shown to perform well on a variety of tasks, and can often achieve state-of-the-art results with relatively little fine-tuning. By starting with a pre-trained model, you can often achieve better performance on your target task more quickly than you would by starting from scratch.

#### 10. What is transfer learning, and how does it work ?

Transfer learning is a machine learning technique that involves using a pre-trained model on one task as a starting point for training a model on a different but related task. The basic idea behind transfer learning is that a model that has been trained on one task can often be re-purposed for another task, either by using the model's learned features directly or by fine-tuning the model on the new task.

The process of transfer learning typically involves several steps:

- Pre-training: The first step is to pre-train a model on a large dataset for a related task. For example, if you are working on a computer vision task, you might pre-train a model on a large dataset of images, such as ImageNet.

- Feature extraction: Once the pre-training is complete, you can use the pre-trained model to extract features from your target dataset. This involves passing each example in your dataset through the pre-trained model and recording the activations of one or more layers in the model. These activations can then be used as features for your target task.

- Fine-tuning: After feature extraction, you can use the extracted features to train a new model on your target task. This typically involves adding one or more new layers on top of the pre-trained model and training the entire model end-to-end on your target task. Alternatively, you can freeze the weights of the pre-trained model and only train the new layers.

#### 11. HOW DO NEURAL NETWORKS LEARN FEATURES ?


Neural networks learn features through a process called backpropagation, which involves computing gradients of the loss function with respect to the weights of the network and updating the weights to minimize the loss.

During training, the input data is passed through the network, and the output is compared to the ground truth labels using a loss function such as cross-entropy. The loss function measures the discrepancy between the predicted output and the ground truth labels.

To update the weights of the network, the gradients of the loss function with respect to the weights are computed using backpropagation. Backpropagation involves computing the gradient of the loss function with respect to each layer of the network, starting from the output layer and working backwards through the network. This process is repeated for each input in the training data set.

Once the gradients have been computed, the weights of the network can be updated using an optimization algorithm such as stochastic gradient descent. The optimization algorithm adjusts the weights in the direction of the negative gradient of the loss function, which results in a decrease in the loss over time.

As the weights of the network are updated, the network learns to extract useful features from the input data that are relevant for predicting the output. The lower layers of the network learn simple features such as edges and curves, while the higher layers learn more complex features that are relevant for the specific task at hand.

#### 12. WHY IS FINE-TUNING BETTER THAN START-UP TRAINING ?


Fine-tuning a pre-trained model is often better than starting from scratch because the pre-trained model has already learned useful features from a large amount of data. Fine-tuning allows us to take advantage of these learned features and adapt them to a new task with relatively small amounts of new data.

Starting from scratch would require training a new model from the beginning with no prior knowledge, which would typically require a much larger amount of data to achieve good performance. Additionally, training a new model from scratch can be time-consuming and computationally expensive.

Fine-tuning a pre-trained model involves taking the pre-trained model and adjusting its weights to better fit the new task. This is typically done by freezing the weights of the earlier layers of the pre-trained model, which contain the general features that are relevant to many tasks, and then training only the later layers on the new data.

Fine-tuning allows us to take advantage of the pre-trained model's ability to extract general features that are relevant to many tasks, while also allowing the model to adapt to the specific features of the new task. This can lead to faster training times and better performance than starting from scratch.
