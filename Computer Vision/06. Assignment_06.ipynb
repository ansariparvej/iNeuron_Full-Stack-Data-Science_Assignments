#### 1. What is the difference between TRAINABLE and NON-TRAINABLE PARAMETERS ?

In machine learning and deep learning, parameters refer to the variables that a model uses to make predictions. These parameters can be divided into two main categories: trainable and non-trainable parameters.

Trainable parameters, also known as learnable parameters, are the variables that a model adjusts during training to improve its performance. These parameters are updated iteratively as the model processes the input data and adjusts its predictions. Examples of trainable parameters include weights and biases in neural networks, coefficients in linear regression models, and decision boundaries in decision trees.

Non-trainable parameters, on the other hand, are fixed variables that do not change during training. These parameters are often set by the user or determined by the architecture of the model. Examples of non-trainable parameters include hyperparameters like learning rate, batch size, and regularization strength, as well as pre-trained embeddings in natural language processing models.

#### 2. In the CNN architecture, where does the DROPOUT LAYER go ? 

During the training phase of a Convolutional Neural Network (CNN) architecture, the dropout layer randomly drops out a specified fraction of the neurons in the previous layer, typically set between 0.1 to 0.5. This means that during each training iteration, a different subset of neurons is dropped out, forcing the network to learn more robust features.

During the testing or inference phase, the dropout layer is typically disabled, and all neurons in the previous layer are used to make predictions. This is because the dropout layer is only used as a regularization technique during training and does not have any effect on the actual predictions made by the network.

The reason for disabling dropout during testing is to ensure that the model's predictions are consistent and reliable. If dropout was enabled during testing, the predictions would be affected by the randomly dropped out neurons, leading to inconsistencies and reducing the accuracy of the model.

In summary, during training, the dropout layer randomly drops out a specified fraction of neurons in the previous layer to prevent overfitting, while during testing or inference, the dropout layer is disabled, and all neurons in the previous layer are used to make predictions.

#### 7. What is NORMALIZATION OF DATA ?

Data normalization is a common preprocessing technique in machine learning that involves rescaling the values of the features in the dataset to a similar range. The goal of normalization is to ensure that each feature contributes equally to the learning process and to improve the convergence rate of the optimization algorithm during training.

There are several different methods for normalizing data, but two of the most commonly used methods are:

==> Min-Max normalization: This technique rescales the feature values to a range between 0 and 1. This is done by subtracting the minimum value of the feature from each value and then dividing by the range of the feature (i.e., the difference between the maximum and minimum values). The resulting values are in the range [0, 1].

==> Z-score normalization: This technique rescales the feature values to have a mean of 0 and a standard deviation of 1. This is done by subtracting the mean of the feature from each value and then dividing by the standard deviation of the feature. The resulting values have a mean of 0 and a standard deviation of 1.

Normalization can help improve the performance of machine learning models, particularly those that use gradient-based optimization algorithms, by preventing the features with larger values from dominating the learning process. It can also help improve the interpretability of the model, as the features are all on a similar scale.

#### 8. What is IMAGE AUGMENTATION and how does it work ?

Image augmentation is a technique used in computer vision and deep learning to increase the size and diversity of a dataset by applying various transformations to existing images. This can help improve the performance of the model by allowing it to learn from more diverse and varied examples of the same object or scene.

The most common types of image augmentations include:

- Rotation: Rotating the image by a certain angle (e.g., 90 degrees).

- Flipping: Flipping the image horizontally or vertically.

- Zooming: Zooming in or out of the image.

- Translation: Shifting the image in a certain direction.

- Shearing: Distorting the image by changing the shape of the object.

- Color jittering: Modifying the brightness, contrast, saturation, or hue of the image.

Image augmentation works by applying these transformations to the original images in the dataset, creating new images that are similar but not identical to the original images. These augmented images can then be used to train the model, effectively increasing the size of the dataset and providing the model with more varied examples to learn from.

By using image augmentation, the model can become more robust to variations in the input data, such as changes in lighting, angle, or position. This can help improve the model's accuracy and generalization ability on new, unseen data.

It's worth noting that not all types of augmentations are appropriate for all types of images or tasks. The specific types and parameters of image augmentations used should be chosen based on the specific problem being solved and the characteristics of the data.

#### 9. What is DECLINE IN LEARNING RATE ?


A decline in learning rate refers to the strategy of reducing the learning rate over time during the training process.

During the training process, the model parameters are updated based on the gradient of the loss function with respect to the parameters. If the learning rate is too high, the parameters may update too quickly, and the optimization process may overshoot the optimal values and fail to converge. On the other hand, if the learning rate is too low, the optimization process may become stuck in a suboptimal solution or take a long time to converge.

To overcome these issues, it is often useful to start with a relatively high learning rate and gradually decrease it over time as the optimization process progresses. This is called a learning rate schedule or learning rate decay.

There are several common strategies for learning rate decay, including:

- Step decay: Reducing the learning rate by a fixed factor after a certain number of epochs or iterations.

- Exponential decay: Reducing the learning rate exponentially over time.

- Cosine annealing: Gradually decreasing the learning rate in a cyclical manner, with a periodicity determined by the total number of epochs.

By reducing the learning rate over time, the optimization process can converge more effectively and efficiently to a better solution. This strategy can be particularly useful for deep neural networks, where the optimization process can be complex and computationally expensive.

#### 10. What does EARLY STOPPING CRITERIA mean ?

Early stopping is a technique used in machine learning to prevent overfitting of a model during training. Overfitting occurs when a model is too complex and has learned the noise in the training data rather than the underlying patterns, leading to poor performance on new, unseen data.

Early stopping criteria refer to the conditions used to determine when to stop the training process to avoid overfitting. Instead of training the model until it converges to the minimum training loss, early stopping stops the training process before the model begins to overfit the training data.

The most common early stopping criteria are based on monitoring the validation loss, which is the loss on a separate validation dataset that is not used during training. The training process is stopped when the validation loss stops improving or starts to increase, indicating that the model has begun to overfit the training data.
