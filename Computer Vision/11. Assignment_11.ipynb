#### 1.	What do REGION PROPOSALS entail?

**Ans:** In object detection tasks, region proposals refer to the process of generating a set of candidate object bounding boxes in an image. The goal of region proposal is to identify areas of the image that are likely to contain an object of interest, and to provide candidate bounding boxes that can be further processed by an object detection algorithm to classify and localize the objects.

Region proposal methods typically use techniques such as sliding windows or selective search to generate a large set of candidate bounding boxes. These methods scan the image at different scales and locations to identify regions that may contain an object. The candidate bounding boxes are typically represented as a set of coordinates that define the location and size of the bounding box.

Once the candidate bounding boxes are generated, they are typically passed to an object detection algorithm, such as a convolutional neural network, to classify and localize the objects within each box. The object detection algorithm uses the candidate bounding boxes to extract features from the image within each box and to make predictions about the presence and location of objects.

#### 2.	What do you mean by NON-MAXIMUM SUPPRESSION? (NMS)

**Ans:** Non-Maximum Suppression (NMS) is a post-processing technique commonly used in object detection to eliminate overlapping bounding boxes that may have been generated by a region proposal algorithm. The goal of NMS is to keep only the most confident bounding boxes and discard the rest.

NMS works by first sorting the candidate bounding boxes according to their confidence scores. Then, starting with the bounding box that has the highest confidence score, it checks the overlap between this box and all other remaining boxes. If the overlap between the two boxes is above a certain threshold (e.g., 0.5), the lower-confidence box is discarded. If the overlap is below the threshold, the box is kept and considered for further overlap checks with other boxes. This process is repeated until all boxes have been checked.

The output of NMS is a set of bounding boxes that are considered to be the most confident and non-overlapping. This helps to reduce the number of false positives and produces a cleaner and more accurate detection result.

To compute mAP, the object detection model generates a set of candidate bounding boxes and assigns a confidence score to each box. The model then compares the predicted boxes to the ground truth boxes to determine the precision and recall of the predictions.

Precision measures the percentage of correct positive predictions among all positive predictions, while recall measures the percentage of correct positive predictions among all ground truth positives. The average precision is the area under the precision-recall curve, which plots the precision against the recall at different IoU thresholds.

#### 5.	What is an IOU (INTERSECTION OVER UNION)?

**Ans:** 

Intersection over Union (IoU) is a commonly used metric in computer vision to evaluate the accuracy of object detection or segmentation models. It measures the overlap between a predicted bounding box (or region) and a ground truth bounding box (or region) for an object in an image.

To calculate IoU, the intersection between the predicted bounding box and the ground truth bounding box is divided by the union of the two boxes. The result is a value between 0 and 1, where a higher value indicates a better overlap between the two boxes.

The IoU metric is used to evaluate the accuracy of object detection or segmentation models by comparing the predicted bounding boxes or regions to the ground truth. A higher IoU indicates a better match between the predicted and ground truth bounding boxes or regions.

#### 6.	Describe the PRECISION-RECALL CURVE (PR CURVE)

**Ans:** 

The Precision-Recall Curve (PR curve) is a graphical representation of the trade-off between precision and recall for a binary classification model. It is commonly used to evaluate the performance of machine learning algorithms in tasks such as information retrieval and object detection.

Precision is a metric that measures the proportion of true positive predictions among all positive predictions. In other words, it measures how accurate the positive predictions are. Recall, on the other hand, measures the proportion of true positive predictions among all actual positive examples. In other words, it measures how many of the positive examples are correctly identified by the model.

The PR curve is created by plotting the precision values against the recall values for different thresholds used by the classification model to predict the positive and negative classes. Each point on the curve represents a specific threshold, and the curve itself is a visualization of the precision and recall trade-offs for different values of the threshold.

In a perfect classification scenario, a PR curve would be a step function with precision equal to one at every recall value. However, in real-world scenarios, the curve may be jagged, and the area under the curve (AUC) can be used to evaluate the performance of the classification model. A higher AUC indicates a better trade-off between precision and recall and therefore a better performing model.

The PR curve is a useful evaluation tool for binary classification tasks, particularly when the class distribution is imbalanced, meaning that one class has significantly more examples than the other. In such cases, the PR curve provides a more informative evaluation than the commonly used Receiver Operating Characteristic (ROC) curve, which can be biased towards the majority class.

#### 7.	What is the term "selective search"?

**Ans:** Selective Search is an object proposal algorithm used in computer vision and image processing. It is a popular method for generating region proposals or candidate object regions in an image for further analysis, such as object detection, segmentation, and recognition.

The main advantage of Selective Search over other object proposal algorithms is its ability to generate a large number of high-quality region proposals while being relatively computationally efficient. #### 8.	Describe the R-CNN model's four components.

**Ans:** 

R-CNN (Region-based Convolutional Neural Network) is a popular object detection model that consists of four main components:

- Region Proposal: The first component of R-CNN is to generate region proposals. This is achieved using a selective search algorithm that proposes a set of candidate object regions in an image. These regions are potential locations of objects in the image, and they are used to focus the subsequent steps of the model only on relevant areas of the image.

- Feature Extraction: Once the candidate regions are generated, the second component of the model is to extract features from these regions. This is done by applying a deep convolutional neural network (CNN) to each region. The CNN extracts a fixed-length feature vector from each region, which captures the spatial characteristics of the region and its contents.

- Object Classification: The third component of the model is to classify each region into one of the predefined object categories. This is done by training a linear support vector machine (SVM) on top of the features extracted by the CNN. The SVM takes the feature vector of a region as input and predicts the probability of the region containing an object of a specific class.

- Object Localization: The fourth and final component of the model is to refine the object localization by predicting the coordinates of the bounding box around the object within each region. This is done by training a regression model on top of the features extracted by the CNN. The regression model takes the feature vector of a region as input and predicts the offset values of the bounding box relative to the region's original proposal.

#### 9.	What exactly is the Localization Module?

**Ans:** The Localization Module is a part of the object detection models that is responsible for predicting the bounding boxes around the objects in an image. In most object detection models, the Localization Module is implemented as a regression layer that takes the features extracted by the convolutional layers and outputs the coordinates of the bounding boxes.

The Localization Module can take various forms depending on the object detection model. In some models, such as the Faster R-CNN, the Localization Module is implemented as a Region Proposal Network (RPN) that generates a set of candidate object proposals based on the features extracted from the image. The RPN then refines the proposals by predicting the offsets to the anchors used to generate the proposals.

#### 10. What are the R-CNN DISADVANTAGES?

**Ans:** Some of the main disadvantages of R-CNN include:

- Slow inference speed: R-CNN requires the extraction of features from each region proposal using a deep convolutional network, which can be computationally expensive. This results in slow inference times, which can limit the applicability of the model in real-time applications.

- High memory usage: R-CNN requires the storage of the features extracted from each region proposal, which can consume a large amount of memory. This can limit the size of the batch of region proposals that can be processed at once, which can further increase the inference time.

- Training is slow: Training R-CNN can be slow due to the need to extract features from each region proposal and the large number of parameters in the model.

- Fixed-sized region proposals: R-CNN uses fixed-sized region proposals that are generated by selective search. This can lead to suboptimal results, as objects of different sizes and aspect ratios may not be properly represented by the region proposals.

- Difficulty in fine-tuning: R-CNN uses pre-trained convolutional networks to extract features, which can make it difficult to fine-tune the model on new data or for new tasks.
