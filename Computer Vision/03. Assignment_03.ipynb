#### 1. After each stride-2 conv, why do we double the number of filters?

In a typical convolutional neural network (CNN), it is common to double the number of filters after each stride-2 convolutional layer. This is done to increase the depth of the network and enable it to learn more complex and abstract features from the input data.

When a stride-2 convolutional layer is applied to an input image, it reduces the spatial dimensions of the output by a factor of 2. This means that the number of features that can be learned by the network is also reduced. By doubling the number of filters after each stride-2 convolutional layer, the network can compensate for this reduction in the number of features by learning more diverse and complex features in each layer.

Additionally, increasing the number of filters after each stride-2 convolutional layer can help to prevent overfitting of the model. When the number of filters is too low, the model may not have enough capacity to capture all of the important features of the input data, which can lead to underfitting. On the other hand, if the number of filters is too high, the model may memorize the training data too well and perform poorly on new, unseen data (overfitting). By gradually increasing the number of filters after each stride-2 convolutional layer, the network can strike a balance between underfitting and overfitting, resulting in better generalization performance.

#### 3. What data is saved by Activation Stats for each layer ?


Activation stats are typically used to monitor the behavior of a neural network during training, and they can be used to diagnose and prevent issues such as vanishing gradients or exploding activations. Activation stats are a summary of the activations produced by a layer of the neural network during a forward pass, and they typically include the following data for each layer:

--> Mean: The mean activation value produced by the layer during a forward pass.

--> Standard deviation: The standard deviation of the activation values produced by the layer during a forward pass.

--> Minimum: The minimum activation value produced by the layer during a forward pass.

--> Maximum: The maximum activation value produced by the layer during a forward pass.

--> Histogram: A histogram of the activation values produced by the layer during a forward pass.

Collecting these statistics for each layer of the neural network can help to identify issues such as dead neurons (neurons that produce zero activations), saturation (neurons that produce the same activation value regardless of the input), or exploding activations (neurons that produce very large activation values that can cause numerical instability during training). These statistics can also be used to visualize the distribution of activations produced by each layer, which can provide insights into the behavior of the network and help to diagnose and resolve issues during training.

#### 5. What are the drawbacks of activations above zero ?

Activations above zero can have some drawbacks, including:

--> Vanishing gradients: When activation values become very large, they can saturate the activation functions used in the neural network, which can cause the gradient of the loss function to become very small during backpropagation. This can make it difficult for the network to update the weights properly and can lead to slow or stagnant learning.

--> Exploding gradients: Similarly, if activation values become too large, the gradient of the loss function can become very large during backpropagation, which can lead to numerical instability and cause the network to fail to converge or produce NaN values.

--> Overfitting: If activation values are allowed to become too large, the network can become too complex and may overfit to the training data, producing high accuracy on the training set but poor generalization to new data.

There are several benefits and drawbacks to practicing in larger batches in deep learning:

--> Faster training: Larger batches can be processed more quickly on modern hardware, as matrix operations are often optimized for larger batch sizes.
--> More stable gradients: Larger batches can result in more stable and consistent gradient updates, which can help to prevent the model from getting stuck in local optima.

**Drawbacks:**

--> Poorer convergence: In some cases, larger batch sizes can lead to slower convergence, as the model may not be able to learn as quickly from larger updates.
--> Reduced learning rate sensitivity: Larger batches may require higher learning rates to achieve good performance, which can make the model more sensitive to the learning rate schedule.

#### 7. Why should we avoid starting training with a high learning rate ?

Starting training with a high learning rate can cause several issues that may adversely affect the performance of the model. Here are some reasons why it's not recommended to start training with a high learning rate:

--> Unstable Training: Starting with a high learning rate can cause the weights to be updated too aggressively, leading to oscillations or instability during training. This can cause the loss to fluctuate wildly, making it difficult to optimize the model.

--> Poor Convergence: If the learning rate is too high, it may cause the optimization algorithm to overshoot the global minimum of the loss function and miss it altogether. As a result, the model may fail to converge to an optimal solution or may take much longer to converge.

--> Divergence: In some cases, starting with a high learning rate can cause the optimization algorithm to diverge, leading to exploding gradients and NaNs (not-a-number) in the loss function. This can happen when the learning rate is so high that the weights are updated in such a way that the loss function becomes undefined.

--> Inefficient training: When the learning rate is too high, the optimization algorithm will make large jumps in the parameter space, potentially skipping over good local optima that could be reached with smaller steps. This can make the training process inefficient and result in suboptimal performance.
