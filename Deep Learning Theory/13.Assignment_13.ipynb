It is generally preferable to use a Logistic Regression classifier rather than a classical Perceptron because Logistic Regression has several advantages over the Perceptron:

- Logistic Regression produces probabilistic outputs: Logistic Regression produces outputs that are continuous values between 0 and 1, which can be interpreted as probabilities. This makes it easier to interpret the outputs of the model and to apply the model in settings where probabilistic predictions are needed, such as in classification with imbalanced classes or in ranking tasks.

- Logistic Regression has a well-defined probabilistic interpretation: Logistic Regression can be interpreted as a linear model that is transformed by the logistic function, which provides a well-defined probabilistic interpretation of the model. This allows us to use techniques such as maximum likelihood estimation and Bayesian inference to estimate the parameters of the model and to perform model selection.

- Logistic Regression can handle non-linearly separable data: Unlike the Perceptron, which can only learn linearly separable decision boundaries, Logistic Regression can learn non-linear decision boundaries by using non-linear transformations of the input features.

To make a Perceptron equivalent to a Logistic Regression classifier, we can apply a simple tweak known as the Perceptron trick. The Perceptron trick involves adding a bias term to the Perceptron and modifying the update rule to include the bias term as well. Specifically, the Perceptron trick involves updating the weights and bias as follows:

w ← w + αyixi

b ← b + αyi

where w is the weight vector, b is the bias term, α is the learning rate, xi is the input vector, yi is the target output (either -1 or 1), and the update is only applied when the target output and predicted output have different signs.

By adding the bias term and modifying the update rule, we can train a single-layer Perceptron to produce outputs that are equivalent to the outputs of a Logistic Regression classifier. However, unlike Logistic Regression, the Perceptron will only produce binary outputs (either -1 or 1) and will not provide probabilistic predictions.

#### 2.	Why was the logistic activation function a key ingredient in training the first MLPs?

**Ans:**

The logistic activation function, also known as the sigmoid function, was a key ingredient in training the first Multilayer Perceptrons (MLPs) because it is a smooth, differentiable function that can be used to represent non-linear transformations of the input data.

The sigmoid function takes as input a real-valued number and maps it to a value between 0 and 1, which can be interpreted as a probability. This property of the sigmoid function was important for training MLPs to perform binary classification tasks, where the goal is to predict whether an input belongs to one class or the other.

In MLPs, the output of each neuron is calculated as a weighted sum of the inputs, followed by the application of an activation function. By using the sigmoid activation function, MLPs were able to learn non-linear decision boundaries that were not possible with linear models such as the Perceptron. The smooth, differentiable nature of the sigmoid function also made it possible to use backpropagation to train the weights of the MLPs, which is an efficient algorithm for computing the gradients of the loss function with respect to the model parameters.

Three popular activation functions used in deep learning are:

- Rectified Linear Unit (ReLU): The ReLU function is defined as f(x) = max(0, x), and returns the input if it is positive, and zero otherwise. The ReLU function has been shown to be effective in training deep neural networks.

- Tanh: The hyperbolic tangent function, or tanh, is defined as f(x) = (e^(2x) - 1) / (e^(2x) + 1), and has a characteristic S-shaped curve that maps the input to a value between -1 and 1.

#### 6.	What is backpropagation and how does it work? What is the difference between backpropagation and reverse-mode autodiff?

**Ans:**

Backpropagation is an algorithm used to calculate the gradients of the loss function with respect to the weights and biases of a neural network, which is necessary for training the network using gradient-based optimization algorithms. It works by propagating the error from the output layer back through the network to update the weights and biases in each layer, using the chain rule of differentiation.

Here's a brief overview of how backpropagation works:

- Forward Pass: The input is propagated forward through the network, calculating the output of each neuron in each layer, until the final output is produced.

- Backward Pass: The error between the predicted output and the actual output is calculated, and the error is propagated backwards through the network using the chain rule to calculate the gradient of the loss function with respect to the weights and biases.

- Weight Update: The weights and biases are updated using the calculated gradients, using a gradient descent algorithm or a variant thereof, to minimize the loss function.

Reverse-mode autodiff, also known as backpropagation through time (BPTT), is a general method for calculating gradients in computational graphs that involves multiple input and output nodes. It works by recursively applying the chain rule of differentiation from the output nodes to the input nodes, accumulating the gradients along the way.

The main difference between backpropagation and reverse-mode autodiff is that backpropagation is specifically designed for computing gradients in neural networks, which have a specific structure with layers of interconnected neurons, while reverse-mode autodiff is a more general method that can be applied to any computational graph. However, both methods ultimately rely on the chain rule of differentiation to compute the gradients of the output with respect to the inputs.

#### 7.	Can you list all the hyperparameters you can tweak in an MLP? If the MLP overfits the training data, how could you tweak these hyperparameters to try to solve the problem?

**Ans:**

Here are some of the hyperparameters that can be tweaked in an MLP:

- Number of hidden layers: This refers to the number of layers between the input and output layers.

- Number of neurons in each hidden layer: This refers to the number of artificial neurons in each hidden layer.

- Activation functions: The activation function determines the output of each neuron in the network.

- Learning rate: The learning rate controls how quickly the weights of the network are updated during training.

- Batch size: The batch size determines the number of samples that are processed at once during training.

- Number of epochs: This refers to the number of times the entire training dataset is passed through the network during training.

- Regularization: Regularization techniques, such as L1 and L2 regularization, can be used to prevent overfitting.

- Dropout: Dropout is a regularization technique that randomly drops out some neurons during training to prevent overfitting.

If an MLP overfits the training data, here are some hyperparameters that could be tweaked to try to solve the problem:

- Decrease the number of neurons in each hidden layer or decrease the number of hidden layers to reduce model complexity.

- Increase the regularization parameter or add dropout to reduce overfitting.

- Decrease the learning rate to prevent the model from overfitting too quickly to the training data.

- Increase the batch size to get a better estimate of the true gradients and reduce overfitting.
