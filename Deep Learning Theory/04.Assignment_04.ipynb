#### 2.	Is TensorFlow a drop-in replacement for NumPy? What are the main differences between the two?

**Ans:** 

While TensorFlow and NumPy share some similarities in terms of their array-based computation capabilities, TensorFlow is not a drop-in replacement for NumPy. Here are some main differences between the two:

- TensorFlow is designed specifically for deep learning and includes a suite of machine learning tools and APIs, while NumPy is a general-purpose library for numerical computing.

- TensorFlow is optimized for GPU computation and distributed computing, while NumPy is primarily designed for CPU-based computation.

- TensorFlow uses a dataflow graph to represent computation, which allows for more efficient execution of complex calculations, while NumPy performs computation using an imperative programming model.

- TensorFlow includes additional features such as automatic differentiation, model checkpointing, and data preprocessing tools, while NumPy focuses primarily on array-based computation.

Yes, besides regular tensors, TensorFlow provides the following data structures:

a mutable tensor that persists across multiple calls to a function. It is often used for model parameters.

- Constant: 
an immutable tensor that does not change during program execution.

- SparseTensor: 
a tensor that stores only nonzero values, along with their indices. This can be more efficient than storing dense tensors when most of the values are zero.
    
- RaggedTensor: 
a tensor with variable numbers of elements along one or more dimensions.

an object representing a sequence of data elements that can be iterated over, typically used for loading large datasets in batches.
