#### 1. Explain the Activation Functions in your own language
- a) sigmoid
- b) tanh
- c) ReLU
- d) ELU
- e) LeakyReLU
- f) swish

**Ans:**


a) Sigmoid: Sigmoid is an activation function that maps any input to a value between 0 and 1. It is a smooth and differentiable function that was popular in early neural networks. It is still used in some applications where a probability-like output is desired.

b) tanh: Tanh is similar to sigmoid but it maps inputs to a value between -1 and 1. It is also smooth and differentiable and was commonly used before ReLU became popular. Tanh is symmetric around the origin, meaning it can output negative or positive values.

c) ReLU: ReLU stands for Rectified Linear Unit, and it is currently one of the most widely used activation functions. It simply sets negative input values to zero and keeps positive values as they are. ReLU is fast and easy to compute and also helps avoid the vanishing gradient problem.

d) ELU: ELU stands for Exponential Linear Unit, and it is similar to ReLU but it allows for negative values. ELU is smooth and differentiable like sigmoid and tanh but is faster than these functions. It has shown better performance than ReLU in some cases.

e) LeakyReLU: LeakyReLU is a variant of ReLU that allows a small, non-zero gradient when the input is negative. This helps to avoid the "dying ReLU" problem where certain neurons become inactive and never contribute to the network's output.

f) Swish: Swish is a relatively new activation function that has shown promising results in some cases. It is similar to ReLU but has a smoother curve that saturates at large input values. Swish is also differentiable and computationally efficient like ReLU.

#### 2. What happens when you increase or decrease the optimizer learning rate?

**Ans:**

The learning rate is a hyperparameter that determines the step size at each iteration while moving toward a minimum of a loss function during optimization. If the learning rate is too low, the model may take a long time to converge or may get stuck in a suboptimal local minimum. On the other hand, if the learning rate is too high, the optimizer may overshoot the minimum and diverge or oscillate around it.

Therefore, increasing the learning rate may cause the model to converge faster, but may also increase the risk of overshooting the minimum and make the optimization unstable. On the other hand, decreasing the learning rate may make the optimization more stable and less likely to overshoot the minimum, but may also slow down the convergence of the model. The optimal learning rate usually depends on the specific problem and the characteristics of the data and the model.

#### 3. What happens when you increase the number of internal hidden neurons?

**Ans:**

When we increase the number of internal hidden neurons in a neural network, the model's capacity to learn complex patterns increases, and the model becomes more expressive. However, adding too many neurons can lead to overfitting. In other words, the model may start to memorize the training data rather than learning to generalize to new data. Moreover, increasing the number of hidden neurons may also increase the training time and memory requirements of the model. Therefore, increasing the number of hidden neurons should be done judiciously and with proper regularization techniques.

#### 4. What happens when you increase the size of batch computation?

**Ans:**

Increasing the size of batch computation generally results in faster training time due to the more efficient use of hardware resources. This is because the computations are done in parallel, which can utilize the GPU more effectively. Additionally, increasing the batch size can lead to more accurate gradient estimates, which may lead to faster convergence and better generalization performance. However, using larger batch sizes can also lead to slower convergence in some cases and may require a larger learning rate to compensate for the increased noise in the gradient estimates. Furthermore, using larger batch sizes requires more memory, which can be a concern when working with limited memory resources.

#### 5. Why we adopt regularization to avoid overfitting?

**Ans:**

We adopt regularization techniques to avoid overfitting because overfitting occurs when a model becomes too complex and learns to fit the noise and idiosyncrasies of the training data, rather than the underlying patterns that generalize to new data. Regularization techniques aim to reduce the model's complexity and constrain the weights to take on smaller values, making it harder for the model to overfit the training data. By doing so, regularization helps to improve the model's ability to generalize to new, unseen data, leading to better performance on the test set.

#### 6. What are loss and cost functions in deep learning?

**Ans:**

In deep learning, a loss function (also known as a cost function or objective function) measures the difference between the predicted output and the actual output (or target) for a given input example. The goal is to minimize the value of the loss function during training, which leads to better predictions on new examples.

The choice of loss function depends on the type of problem being solved. For example, for regression problems, commonly used loss functions include mean squared error (MSE) and mean absolute error (MAE), while for classification problems, cross-entropy loss is often used. Other loss functions, such as hinge loss and triplet loss, are used for specific tasks such as binary classification and face recognition, respectively.

The cost function, on the other hand, is the average of the loss functions for all the training examples. The goal of training a neural network is to minimize the cost function by adjusting the network's parameters (weights and biases) using an optimization algorithm such as stochastic gradient descent (SGD). Minimizing the cost function leads to a model that can generalize well to new examples.

#### 7. What do you mean by underfitting in neural networks?

**Ans:**
Underfitting in neural networks refers to a situation where the model is not complex enough to capture the underlying patterns in the data. As a result, the model may perform poorly not only on the training data but also on unseen test data. When a model is underfitting, it may be because the model is too simple or the training dataset is not diverse enough to capture all possible variations in the data. Underfitting can be addressed by increasing the model's capacity or complexity, collecting more data, or using a different model architecture.

#### 8. Why we use Dropout in Neural Networks?

**Ans:**

Dropout is a regularization technique used in neural networks to prevent overfitting. Overfitting occurs when a model is trained to fit the training data so closely that it performs poorly on new, unseen data. Dropout addresses this problem by randomly dropping out (setting to zero) some of the neurons in the neural network during each training step. This makes it harder for the network to memorize the training data and encourages it to learn more robust features that generalize better to new data.

The effect of dropout is similar to training an ensemble of many different neural networks, each with a different subset of neurons, and then averaging their predictions. This helps prevent overfitting and improves the generalization performance of the model. Dropout is typically applied to the hidden layers of a neural network, but it can also be applied to the input or output layers in some cases.
