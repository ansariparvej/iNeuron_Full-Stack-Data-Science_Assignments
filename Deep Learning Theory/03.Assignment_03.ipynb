#### 1.	Is it OK to initialize all the weights to the same value as long as that value is selected randomly using He initialization?

**Ans:**


No, it is not recommended to initialize all the weights to the same value, even if that value is selected randomly using He initialization. This is because initializing all weights to the same value means that all neurons in the same layer will be computing the same function, and this can lead to the symmetry problem in neural networks. This can make the network more difficult to train, as it can result in vanishing gradients and slow convergence.

He initialization is a method of initializing the weights in a neural network that takes into account the number of inputs to each neuron. This helps to prevent vanishing and exploding gradients during training. However, it is still important to initialize the weights randomly to avoid symmetry and ensure that the network can learn different features from the input data.

Therefore, it is recommended to initialize the weights randomly using a method such as He initialization, Xavier initialization, or Glorot initialization, which take into account the number of inputs and outputs of each neuron. This can help ensure that the network can learn a diverse range of features from the input data and improve its ability to generalize to new data.

#### 2.	Is it OK to initialize the bias terms to 0?

**Ans:**
Yes, it is generally okay to initialize the bias terms to 0. Unlike weights, which control the strength of the connections between neurons, the bias terms only shift the activation function of each neuron, and so their initialization is less critical. Initializing bias terms to 0 does not introduce the same kind of symmetry problems that initializing weights to the same value can cause.

That being said, there are other initialization methods for bias terms that can be used, such as initializing them to a small positive or negative constant value, or using a random initialization method similar to that used for weights.

#### 3.	Name three advantages of the SELU activation function over ReLU.

**Ans:**


The Scaled Exponential Linear Unit (SELU) activation function is a self-normalizing activation function that is becoming increasingly popular in deep learning. Some advantages of the SELU activation function over the Rectified Linear Unit (ReLU) activation function are:

- Self-normalizing property: 

The SELU activation function is designed to have a mean activation close to 0 and a standard deviation close to 1 during training. This helps to prevent the vanishing and exploding gradients problems that can occur in deep networks, and can lead to faster and more stable training.

- Improved performance on deep networks: 

Because of its self-normalizing property, the SELU activation function has been shown to work well on deep neural networks with many layers. This is in contrast to the ReLU activation function, which can suffer from the vanishing gradients problem on deep networks.

- Continuous and differentiable: 

The SELU activation function is continuous and differentiable, which makes it easy to use in standard optimization algorithms such as gradient descent. This is in contrast to other activation functions, such as the step function, which are not differentiable and can make optimization more difficult.


#### 4.	In which cases would you want to use each of the following activation functions: SELU, leaky ReLU (and its variants), ReLU, tanh, logistic, and softmax?

**Ans:** 

Different activation functions are suitable for different types of neural network architectures and tasks. Here are some general guidelines for when to use different activation functions:

- SELU: 
The Scaled Exponential Linear Unit (SELU) activation function is a self-normalizing activation function that is useful for deep neural networks with many layers. It can help to prevent the vanishing and exploding gradients problems that can occur in deep networks.

- Leaky ReLU and its variants: 
Leaky ReLU and its variants (e.g., Parametric ReLU, Randomized ReLU) are useful when ReLU fails to produce the desired result. These variants of the ReLU function allow for a small, non-zero gradient when the input is negative, which can help to prevent the dying ReLU problem. In general, if you experience the vanishing gradient problem, it can be useful to experiment with different variants of ReLU to see which one works best for your problem.

- ReLU: 
Rectified Linear Units (ReLU) are commonly used as activation functions in neural networks. ReLU is often a good choice for deep neural networks because it is fast to compute and does not suffer from the vanishing gradients problem.

- tanh: 
The hyperbolic tangent (tanh) function is similar to the sigmoid function, but is symmetric around zero. It is often used in neural networks for binary classification problems, such as predicting whether an image contains a cat or a dog.

- logistic: 
The logistic function is similar to the tanh function and is often used in binary classification problems. It is not commonly used in deep neural networks due to its vanishing gradients problem, but it can be useful for simpler models.

- softmax: 
The softmax function is commonly used in the output layer of neural networks for multi-class classification problems. It produces a probability distribution over the possible classes and can be used to make predictions for new inputs.

#### 5.	What may happen if you set the momentum hyperparameter too close to 1 (e.g., 0.99999) when using an SGD optimizer?

**Ans:** 

Setting the momentum hyperparameter too close to 1 (e.g., 0.99999) when using a Stochastic Gradient Descent (SGD) optimizer may cause the optimizer to overshoot the minimum of the cost function, causing it to oscillate around the minimum rather than converging to it. This is because the momentum term becomes dominant and the optimizer tends to continue moving in the same direction as the previous updates with a larger and larger step. As a result, the optimization process becomes unstable and takes longer to converge. In some cases, the optimizer may not converge at all, and the loss function may start to increase instead of decreasing. Therefore, it is important to choose an appropriate momentum value for the problem at hand, typically in the range of 0.8 to 0.9.

#### 6.	Name three ways you can produce a sparse model.

**Ans:**

Sparse models are models that have many weights close to zero. There are several ways to produce a sparse model:

- L1 regularization: 
This is a common technique for inducing sparsity in deep learning models. L1 regularization adds a penalty term to the loss function that is proportional to the absolute value of the weights, which encourages the optimizer to set many of them to zero.

- Dropout: 
Dropout is a regularization technique that randomly drops out a fraction of the neurons in a layer during training. This forces the remaining neurons to learn more robust features, which can lead to sparser models.

- Weight pruning: 
Weight pruning is a technique that involves removing small-weight connections from the model after training. This can be done in several ways, such as removing all connections below a certain threshold, or using more sophisticated methods such as magnitude-based pruning or iterative pruning.

#### 7.	Does dropout slow down training? Does it slow down inference (i.e., making predictions on new instances)? What about MC Dropout?

**Ans:**

Yes, dropout can slow down training because it introduces additional computations during each forward and backward pass. During training, dropout randomly drops out a fraction of the neurons, and the remaining neurons are scaled up by a factor of 1/(1 - dropout rate) to preserve their expected values. This scaling up operation requires additional computations during training, which can slow down the process.

However, during inference (i.e., making predictions on new instances), dropout does not slow down the process because all neurons are used and no scaling operation is needed. The dropout layers are simply removed from the network, and the remaining weights are scaled by the retention probability (1 - dropout rate) to preserve their expected values.

MC Dropout (Monte Carlo Dropout) is a variant of dropout that can be used during inference to obtain more accurate predictions and estimate the uncertainty of the model. During inference, MC Dropout applies dropout multiple times with different dropout masks to produce a distribution of predictions. This can be computationally expensive because it requires multiple forward passes, but it can provide more robust predictions and estimate the model uncertainty, which can be useful in certain applications such as active learning or safety-critical systems.
